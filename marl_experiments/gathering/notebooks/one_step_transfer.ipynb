```python
%matplotlib inline
import os 
import os.path as osp

import numpy as np
import matplotlib.pylab as plt
import pandas as pd 

from marl.utils import tb_utils
```

```python
PATH = "/scratch/wellman_root/wellman1/mxsmith/results/marl/gathering/2022_11_29_one_step_transfer"
```

## Best Response to Random

### Learner

```python
df = tb_utils.events_to_dataframe(osp.join(PATH, "br_learner"))
print(df.columns)
```

```python
fig, axes = plt.subplots()

axes.plot(df["learner_step"], df["loss"])
axes.set_ylabel("Loss")
axes.set_xlabel("Learner Step")

plt.show()


fig, axes = plt.subplots()

axes.plot(df["learner_step"], df["param_updates_norm"])
axes.set_ylabel("Parameter Update Norm")
axes.set_xlabel("Learner Step")

plt.show()
```

### Training Arenas

```python
train_dfs = []
for i in range(4):
    train_dfs.append(tb_utils.events_to_dataframe(osp.join(PATH, f"br_agent_train_arena_{i}")))
print(train_dfs[0].columns)
```

```python
fig, axes = plt.subplots()

for i in range(4):
    axes.plot(train_dfs[i]["learner_step"], train_dfs[i]["train_return_per_episode/0"])

plt.show()
```
