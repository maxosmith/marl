```python
import cloudpickle 
import numpy as np
import tree

from marl.games import gathering
```

```python
path = "/scratch/wellman_root/wellman1/mxsmith/tests/sample.pb"
```

```python
data = cloudpickle.load(open(path, "rb"))
# Data is stored as unwritable DeviceArrays with device axis.
# data = tree.map_structure(lambda x: np.squeeze(np.asarray(x).copy(), axis=0), data)
data = tree.map_structure(lambda x: np.asarray(x).copy(), data)
```

```python
observation = data.data.observation             # (B, S, W, H, E).
action = data.data.action                       # (B, S).
reward = data.data.reward                       # (B, S).
start_of_episode = data.data.start_of_episode   # (B, S).
end_of_episode = data.data.end_of_episode       # (B, S).
extras = data.data.extras                       # Tree.
```

```python
flat_observation = observation.copy()
flat_observation[..., 0] *= 1
flat_observation[..., 1] *= 2
flat_observation[..., 2] *= 3
flat_observation[..., 3] *= 4
flat_observation[..., 4] *= 0
flat_observation = np.sum(flat_observation, axis=-1)
```

```python
B = 20

for s in range(flat_observation.shape[1]):
    print(flat_observation[B, s][:, ::-1].T)
    print(gathering.GatheringActions(action[B, s]))
    print(reward[B, s])
    print(start_of_episode[B, s], end_of_episode[B, s])
    print("\n")
```
